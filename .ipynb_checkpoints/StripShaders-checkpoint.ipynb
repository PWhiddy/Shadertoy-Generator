```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
os.chdir("./allshaders")
```

```python
files = os.listdir("./")
```

```python
sizes = []
for f in files:
    stats = os.stat(f)
    sizes.append(stats.st_size)
    #print(stats.st_size)
```

```python
sizes = sorted(sizes)
```

```python
trim = 1200
n, bins, patches = plt.hist(sizes[:len(sizes)-trim], 80, density=False, facecolor='g', alpha=0.75)

#plt.yscale('log')
plt.xlabel('Size in bytes')
plt.ylabel('Number of Shaders')
plt.title('Distribution of shader sizes on shadertoy (' + str(trim) + ' Largest removed)')
#plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

```python
masses = []
for (b,nz) in zip(bins,n):
    masses.append(nz*(b-170*0.5))
```

```python
# Total storage occupied by shaders of different sizes
plt.plot((bins-170*0.5)[:len(bins)-1], masses)
```

```python
trim = 10
n, bins, patches = plt.hist(sizes[:len(sizes)-trim], 160, density=False, facecolor='g', alpha=0.75)

plt.yscale('log')
plt.xlabel('Size in bytes')
plt.ylabel('Number of Shaders')
plt.title('Distribution of shader sizes on shadertoy (' + str(trim) + ' Largest removed)')
#plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```
